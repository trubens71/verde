```python
import pandas as pd
```

```python
df = pd.read_excel('AFR_T41.xlsx', sheet_name='Regions')
df
```

```python
df = df.melt(id_vars=['Region code','Region name'],
        var_name='Setting',
        value_name='Gross Current Expenditure (x1000)')
df
```

```python
df_total_exp = df.groupby('Region code').sum('Gross Current Expenditure (x1000)')
df_total_exp.rename(columns={'Gross Current Expenditure (x1000)': 'Gross Current Expenditure (x1000) Region Total'}, inplace=True)
df_total_exp
```

```python
df = df.merge(df_total_exp, on='Region code')
df
```

```python
df['Gross Current Expenditure Proportion'] = \
    df['Gross Current Expenditure (x1000)'] / df['Gross Current Expenditure (x1000) Region Total']
df.drop(columns='Gross Current Expenditure (x1000) Region Total', inplace=True)
df
```

```python
df.groupby('Region code').sum('Gross Current Expenditure Proportion')
```

```python
df.to_csv('AFR_T41_region_tidy.csv', index=False)
```
